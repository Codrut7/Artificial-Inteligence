```python
import numpy as np
import pandas as ps
```

```python
df = ps.read_csv('Linear Regression - Sheet1.csv')
df
```

```python
y = df['Y'].as_matrix()
df = df.drop(['Y'],axis=1)
```

```python
x = df.as_matrix()
x = x[0:290]
x = np.reshape(x,(len(x),1))
y = y[0:290]
y = np.reshape(y,(len(y),1))
x = x / np.linalg.norm(x)
ones = np.ones((len(x),1))
x = np.concatenate((x, ones), axis=1)
y = y / np.linalg.norm(y)
```

```python
print(x)
```

```python
def train():
    learning_rate = 0.1
    iterations = 10000
    weights = np.random.randn(2,1)
    for i in range(iterations):
        z = np.dot(x,weights)
        error = 1/(2 * len(x)) * np.sum((z-y) * (z-y))
        dcost_dh = 1/len(x) * (z-y)
        dcost_dw = np.sum(np.dot(x.T,dcost_dh))
        weights -= learning_rate * dcost_dw
        if i % 100 == 0:
            print(error)
    return weights
```

```python
w = train()
```
