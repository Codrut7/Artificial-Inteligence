```python
import pandas as ps
import numpy as np
```

```python
df = ps.read_csv('Salary_Data.csv')
df
```

```python
# Get input and output vectors
x = df['YearsExperience'].as_matrix()
y = df['Salary'].as_matrix()
```

```python
#Normalize the stuff
x = x / np.linalg.norm(x)
y = y / np.linalg.norm(y)

#Reshape

x = np.reshape(x,(len(x),1))
y = np.reshape(y,(len(y),1))

ones = np.ones((len(x),1))

x_bias = np.concatenate((x,ones),axis=1)
```

```python
def train1():
    learning_rate = 0.01
    iterations = 10000
    weights = np.random.randn(2,1)
    for i in range(iterations):
        z = np.dot(x_bias,weights)
        cost = 1/(2) * np.sum((z-y) * (z-y))
        dcost_dz = (z - y)
        dcost_dw = np.dot(x_bias.T,dcost_dz)
        weights -= dcost_dw * learning_rate
        if i % 100 == 0:
            print(cost)
    return weights
```

```python
train1()
```

```python
def train2():
    learning_rate = 0.01
    iterations = 10000
    weights = np.random.randn(1,1)
    bias = 1
    for i in range(iterations):
        z = np.dot(x,weights)
        cost = 1/2  * np.sum((z-y) * (z-y))
        dcost_dz =  (z - y)
        dcost_dw = np.dot(x.T,dcost_dz)
        weights -= learning_rate * dcost_dw
        if i % 100 == 0:
            print(cost)
    return weights,bias
```

```python
train2()
```
