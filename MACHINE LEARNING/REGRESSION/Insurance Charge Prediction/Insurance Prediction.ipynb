```python
import numpy as np
import pandas as ps
```

```python
df = ps.read_csv("insurance.csv")
df
```

```python
# Transform string columns into nubmers (categories)
df['sex'] = df['sex'].astype('category')
df['sex'] = df['sex'].cat.codes
df['smoker'] = df['smoker'].astype('category')
df['smoker'] = df['smoker'].cat.codes
df['region'] = df['region'].astype('category')
df['region'] = df['region'].cat.codes
# Print the dataframe
df
```

```python
y_train = df['charges'].as_matrix()
y_train = np.reshape(y_train,(len(y_train),1))
y_train
```

```python
# Drop the output column from the dataframe
df = df.drop(['charges'],axis=1)
df
```

```python
x_train = df.as_matrix()
print(x_train)
```

```python
x = x_train / np.linalg.norm(x_train)
y = y_train / np.linalg.norm(y_train)
```

```python
def train():
    learning_rate = 0.1
    iterations = 10000
    weights = np.random.randn(6,1)
    bias = np.ones((len(x),1))
    for i in range(iterations):
        z = np.dot(x,weights) + bias
        error = 1/(2 * len(x)) * (np.sum((z-y)*(z-y)))
        dcost_dh = 1/len(x) * (z-y)
        dcost_dw = np.dot(x.T, dcost_dh)
        dcost_db = dcost_dh
        if i % 100 ==0:
            print(error)
        weights -= learning_rate * dcost_dw
        bias -= learning_rate * dcost_db
    return weights, bias
w, b = train()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)

from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
reg = LinearRegression().fit(x_train, y_train)
reg.score(x_test,y_test)
```
