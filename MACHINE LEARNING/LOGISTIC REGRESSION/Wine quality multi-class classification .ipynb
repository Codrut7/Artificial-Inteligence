```python
import pandas as ps
import numpy as np
```

```python
df = ps.read_csv('winequality-red.csv')
df
```

```python
y = df['quality'].as_matrix()
y = np.reshape(y,(len(y),1))
for i in range(len(y)):
    if y[i] <=4:
        y[i] = 0
    elif y[i] > 3 and y[i] <=6:
        y[i] = 1
    else:
        y[i] = 2
np.set_printoptions(threshold=np.inf)
print(y)
```

```python
y_class1 = np.full((len(y),1), 0, dtype=float)
y_class2 = np.full((len(y),1), 0, dtype=float)
y_class3 = np.full((len(y),1), 0, dtype=float)
```

```python
for i in range(len(y)):
    if y[i] == 0:
        y_class1[i] = 1
    elif y[i] == 1:
        y_class2[i] = 1
    else:
        y_class3[i] = 1
np.set_printoptions(threshold=np.inf)
print(y_class1)
```

```python
df = df.drop(['quality'], axis = 1)
x = df.as_matrix()
```

```python
def logistic_loss(y,y_hat):
    return -np.mean(y * np.log(y_hat) + (1-y) * np.log(1-y_hat))
```

```python
def sigmoid(x):
    return (1 / (1 + np.exp(-x)))
```

```python
def d_sigmoid(x):
    return sigmoid(x) * (1 - sigmoid(x))
```

```python
#Normalize the stuff
x = x / np.linalg.norm(x)
ones = np.ones((len(x),1))
x_bias = np.concatenate((x,ones),axis=1)
```

```python
def learn():
    iterations = 10000
    learning_rate = 0.1
    weights1 = np.random.randn(len(x_bias[0]),1)
    weights2 = np.random.randn(len(x_bias[0]),1)
    weights3 = np.random.randn(len(x_bias[0]),1)
    for i in range(iterations):
        z1 = np.dot(x_bias, weights1)
        z2 = np.dot(x_bias, weights2)
        z3 = np.dot(x_bias, weights3)
        a1 = sigmoid(z1)
        a2 = sigmoid(z2)
        a3 = sigmoid(z3)
        error1 = logistic_loss(y_class1, a1)
        error2 = logistic_loss(y_class2, a2)
        error3 = logistic_loss(y_class3, a3)
        derror1_dh = ( 1 / len(x_bias) ) * (a1 - y_class1)
        derror2_dh = ( 1 / len(x_bias) ) * (a2 - y_class2)
        derror3_dh = ( 1 / len(x_bias) ) * (a3 - y_class3)
        derror1_dw = np.dot(x_bias.T, derror1_dh)
        derror2_dw = np.dot(x_bias.T, derror2_dh)
        derror3_dw = np.dot(x_bias.T, derror3_dh)
        if i % 100 == 0:
            print(error1)
            print(error2)
            print(error3)
            print("-----")
        weights1 -= learning_rate * derror1_dw
        weights2 -= learning_rate * derror2_dw
        weights3 -= learning_rate * derror3_dw
    return weights1, weights2, weights3
```

```python
# Pentru un set de inputs x -> x1,x2 ... functia care returneaza max(h1,h2,h3) ne va da clasa rezultat

learn()
```
