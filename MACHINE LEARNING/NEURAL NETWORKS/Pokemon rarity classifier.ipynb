```python
import numpy as np
import pandas as ps
```

```python
df = ps.read_csv("pokemon.csv")
df = df.drop(['Name', "#"], axis = 1)
df
```

```python
# Transform string columns into nubmers (categories)
df['Type 1'] = df['Type 1'].astype('category')
df['Type 1'] = df['Type 1'].cat.codes

df['Type 2'] = df['Type 2'].astype('category')
df['Type 2'] = df['Type 2'].cat.codes

df['Legendary'] = df['Legendary'].astype('category')
df['Legendary'] = df['Legendary'].cat.codes

df
```

```python
y = df['Legendary'].as_matrix()
df = df.drop(['Legendary'], axis = 1)
y = np.reshape(y,(len(y),1))
x = df.as_matrix()
x = x / np.linalg.norm(x)
bias = np.ones((len(x),1))
x_bias = np.concatenate((x,bias),axis = 1)
```

```python
def sigmoid(x):
    return 1 / ( 1 + np.exp(-x))
def d_sigmoid(x):
    return sigmoid(x) * (1 - sigmoid(x))
```

```python
def learn():
    iterations = 1000000
    learning_rate = 0.01
    w1 = np.random.randn(11,5)
    w2 = np.random.randn(5,1)
    for i in range(iterations):
        # Forward propagation
        z1 = np.dot(x_bias, w1)
        a1 = sigmoid(z1)
        z2 = np.dot(a1, w2)
        a2 = sigmoid(z2)
        error = np.sum((a2 - y) * (a2 - y))
        # Backward propagation
        derror_dpred = a2 - y
        da2_dz2 = d_sigmoid(z2)
        dz2_dw2 = a1
        derror_dw2 = np.dot(dz2_dw2.T, derror_dpred * da2_dz2)
        dz2_da1 = w2
        da1_dz1 = d_sigmoid(z1)
        dz1_dw1 = x_bias
        derror_dw1 = np.dot(dz1_dw1.T, np.dot(derror_dpred * da2_dz2, dz2_da1.T) * da1_dz1)
        
        w2 -= learning_rate * derror_dw2
        w1 -= learning_rate * derror_dw1
        
        if i % 100 == 0:
            print(error)
    return w1,w2
        
```

```python
w1,w2 = learn()
```
